```python
# Feature Engineering

cleaned_df = df.copy()

cleaned_df["Date"] = pd.to_datetime(cleaned_df["Date"], dayfirst=True)

cleaned_df = cleaned_df.sort_values(by=["HomeTeam", "Date"])

cleaned_df["HomeLastMatch"] = cleaned_df.groupby("HomeTeam")["Date"].diff().dt.days.fillna(7) 
cleaned_df["AwayLastMatch"] = cleaned_df.apply(
    lambda row: (row["Date"] - cleaned_df[(cleaned_df["AwayTeam"] == row["AwayTeam"]) & 
                                          (cleaned_df["Date"] < row["Date"])]["Date"].max()).days
                if not cleaned_df[(cleaned_df["AwayTeam"] == row["AwayTeam"]) & 
                                  (cleaned_df["Date"] < row["Date"])].empty else 7, 
    axis=1
)

cleaned_df["HomeGoalDiff"] = cleaned_df["FTHG"] - cleaned_df["FTAG"]
cleaned_df["AwayGoalDiff"] = -cleaned_df["HomeGoalDiff"]

cleaned_df["HomeAvgGoals"] = cleaned_df.groupby("HomeTeam")["FTHG"].transform(lambda x: x.rolling(5, min_periods=1).mean())
cleaned_df["AwayAvgGoals"] = cleaned_df.groupby("AwayTeam")["FTAG"].transform(lambda x: x.rolling(5, min_periods=1).mean())

cleaned_df["HomeGoalsLast5"] = df.groupby("HomeTeam")["FTHG"].rolling(5, min_periods=1).mean().reset_index(0, drop=True)
cleaned_df["AwayGoalsLast5"] = df.groupby("AwayTeam")["FTAG"].rolling(5, min_periods=1).mean().reset_index(0, drop=True)

cleaned_df["HomeWinStreak"] = cleaned_df.groupby("HomeTeam")["FTR"].transform(lambda x: x.eq("H").rolling(5, min_periods=1).sum())
cleaned_df["AwayWinStreak"] = cleaned_df.groupby("AwayTeam")["FTR"].transform(lambda x: x.eq("A").rolling(5, min_periods=1).sum())

cleaned_df["HomeAvgConceded"] = cleaned_df.groupby("HomeTeam")["FTAG"].transform(lambda x: x.rolling(5, min_periods=1).mean())
cleaned_df["AwayAvgConceded"] = cleaned_df.groupby("AwayTeam")["FTHG"].transform(lambda x: x.rolling(5, min_periods=1).mean())

cleaned_df["HomeGoalDiffAvg"] = cleaned_df.groupby("HomeTeam")["HomeGoalDiff"].transform(lambda x: x.rolling(5, min_periods=1).mean())
cleaned_df["AwayGoalDiffAvg"] = cleaned_df.groupby("AwayTeam")["AwayGoalDiff"].transform(lambda x: x.rolling(5, min_periods=1).mean())

cleaned_df.head(10)
```

```python
columns_to_use = ["Date", "HomeTeam", "AwayTeam", "HomeLastMatch", "AwayLastMatch", "HomeGoalDiff", "AwayGoalDiff", "HomeAvgGoals", "AwayAvgGoals", "HomeGoalsLast5", "AwayGoalsLast5", "HomeWinStreak", "AwayWinStreak", "HomeAvgConceded", "AwayAvgConceded", "HomeGoalDiffAvg", "AwayGoalDiffAvg", "FTHG", "FTAG", "FTR", "AvgH", "AvgD", "AvgA", "Avg>2.5", "Avg<2.5"]

for cols in cleaned_df.columns:
    if cols not in columns_to_use:
        cleaned_df = cleaned_df.drop(columns=cols)

cleaned_df.columns
```

```python
cleaned_df = cleaned_df.rename(columns={
    "FTHG": "FullTimeHomeTeamGoals", 
    "FTAG": "FullTimeAwayTeamGoals", 
    "FTR": "FullTimeResult", 
    "AvgH": "MarketAverageHomeWinOdds", 
    "AvgD": "MarketAverageDrawWinOdds", 
    "AvgA": "MarketAverageAwayWinOdds", 
    "Avg>2.5": "MarketAverageOver2.5Goals", 
    "Avg<2.5": "MarketAverageUnder2.5Goals"
    })

cleaned_df
```

```python
cleaned_df["HomeTeam"] = cleaned_df["HomeTeam"].apply(lambda x: team_mapping.get(x))
cleaned_df["AwayTeam"] = cleaned_df["AwayTeam"].apply(lambda x: team_mapping.get(x))
cleaned_df["FullTimeResult"] = cleaned_df["FullTimeResult"].map({"H": 1, "D": 0, "A": -1})

cleaned_df
```
